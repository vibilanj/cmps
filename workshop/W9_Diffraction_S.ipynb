```python
import scipy         # Another numerical library
from scipy import integrate

import matplotlib    # Library used for plotting
import numpy as np   # Numerical library
import matplotlib.pyplot as plt # Plot commands
from scipy.fft import fft2, fftshift

# Define some colors using the RGB format

CF_red = (204/255, 121/255, 167/255)
CF_vermillion = (213/255, 94/255, 0)
CF_orange = (230/255, 159/255, 0)
CF_yellow = (240/255, 228/255, 66/255)
CF_green = (0, 158/255, 115/255)
CF_sky = (86/255, 180/255, 233/255)
CF_blue = (0, 114/255, 178/255)
CF_black = (0, 0, 0)
```

```python
w = 2e-3
h = 4e-3
screen_size = 0.1
nPts = 1000

def ApertureRect(x, y):
    if (abs(x) < w and abs(y) < h):
        return 1
    else:
        return 0
    
vApertureRect = np.vectorize(ApertureRect)

xs = np.mgrid[-screen_size:screen_size:1j * nPts, -screen_size:screen_size:1j * nPts][0]
ys = np.mgrid[-screen_size:screen_size:1j * nPts, -screen_size:screen_size:1j * nPts][1]

E_Inc = vApertureRect(xs, ys)
```

```python
fig, ax = plt.subplots(1, 1)

ax.pcolormesh(xs, ys, E_Inc, cmap="bone")
ax.set_aspect("equal")
```

```python
Diff = fft2(E_Inc)

fig, ax = plt.subplots(1, 1)

ax.pcolormesh(abs(fftshift(Diff)), cmap="bone")
ax.set_aspect("equal")
```

```python
r = 0.25e-2
def Aperture(x, y):
    if np.sqrt(x**2 + y**2) < r:
        if abs(x) < 1e-3:
            return 0
        elif abs(y) < 1e-3:
            return 0
        else:
            return 1
    else:
        return 0
    

vAperture = np.vectorize(Aperture)

xs = np.mgrid[-screen_size:screen_size:1j * nPts, -screen_size:screen_size:1j * nPts][0]
ys = np.mgrid[-screen_size:screen_size:1j * nPts, -screen_size:screen_size:1j * nPts][1]

E_Inc = vAperture(xs, ys)
```

```python
fig, ax = plt.subplots(1, 1)

ax.pcolormesh(xs, ys, E_Inc, cmap="bone")
ax.set_aspect("equal")
```

```python
Diff = fft2(E_Inc)

fig, ax = plt.subplots(1, 1)

ax.pcolormesh(abs(fftshift(Diff)), cmap="bone")
ax.set_aspect("equal")
```
